### Regression Analysis
It refers to the process of determining if exists a functional relationship between two "measures". If there exist a linear relationship then we are talking about linear regression.

The equation of the linear regression is expressed as:
\begin{equation}
y = mx + q
\end{equation}
and the scope is to determine the parameters $m$ and $q$ in a way that minimize the quantity:
\begin{equation}
g(m,q) = \sum_{i=1}^n(mx_i + q - y_i)^2
\end{equation}

Minimizing the equation above with the hessian method, given two series of values $X$ and $Y$ the equation of the line can be rewritten as:
\begin{equation}
\bar{y} = m\bar{x} + q
\end{equation}

Moreover, $m$ and $q$ can be calculated by using the variance and covariance values
\begin{equation}
m = \frac{c_{xy}}{s_x^2}
\end{equation}
\begin{equation}
q = \bar{y} - \frac{c_{xy}}{s_x^2}\bar{x}
\end{equation}

```python
x = np.array([2500,2900,3000,3000,3000,3000,3250,3250,3250,3600,4225,4650,4750,5175,5300])
y = np.array([4650,4650,4700,4750,4775,4775,4800,4950,5050,5100,5100,5150,5175,5250,5300])

cov_xy = np.cov(x,y, bias=True)[0][1]

m = cov_xy/x.var()
q = y.mean() - (cov_xy/x.var())*x.mean()
```

```python
plt.figure(figsize=(7,3))
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.scatter(x,y)
plt.plot(x, x*m + q, color="r")
plt.show()
```

However, this method does not provide information on the degree of approximation of the line that fit the relationship between the two measurements. Therefore, it has been introduced the linear correlation coefficient.
\begin{equation}
r_{xy} = \frac{c_{xy}}{s_xs_y}
\end{equation}
it's range is $[-1, 1]$. When the value is 0 means that the data are statistically uncorrelated, when it is 1 means that the measurements are exactly aligned on the line $y=mx + q$. So, it represents the alignment's degree of the pairs of measurements.

```python
r = cov_xy/(x.std()*y.std())
r
```

This correlation coefficient is also known as Pearson Correlation Coefficient
