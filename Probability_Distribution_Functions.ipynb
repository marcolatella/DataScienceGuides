## Probability Distribution Functions

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

### Probability Mass Function & Cumulative Distribution Function - Discrete
Immagine rolling a die. The probability of each outcome of a dice is $\frac{1}{6}$

```python
dice_prob = [1/6 for _ in range(6)]
outcomes = np.array([1,2,3,4,5,6])
plt.figure(figsize=(7,10))

plt.subplot(4,2,1)
plt.hist(outcomes, density=True, bins=range(1,8), edgecolor='white', linewidth=1.2, align='left')
plt.xlabel("PMF")

plt.subplot(4,2,2)
aaa = stats.uniform.cdf(outcomes, 0, 6)
plt.bar([1,2,3,4,5,6], aaa)
plt.xlabel("CDF")

plt.show()
```

The Cumulative Distribution Function $F(X)$ evaluated at $x$ is the probability that $X$ will take a value less than or equal to $x$. So, $F_X{(x)} = P(X \le x)$
An example.
\begin{equation}
F_X{(4)} = P(X \le 4) = P(X=1) + P(X=2) + P(X=3) + P(X=4)
\end{equation}

### Probability Density Function & Cumulative Distribution Function - Continuous
Immagine to measure the height of 100 women in cm. From the measurements we calculate the mean and the standard deviation. mean=165, std=10.

```python
plt.figure(figsize=(7,10))

# simulate a measurement of 100 women height, in a way that
x = np.sort(np.random.normal(165, 10, 100))

plt.subplot(4,2,3)
plt.hist(x)
plt.xlabel("Frequency")

plt.subplot(4,2,1)
height_dist = stats.norm.pdf(x, 165, 10)
plt.plot(x, height_dist)
plt.xlabel("PDF")

plt.subplot(4,2,2)
cdf = stats.norm.cdf(x, 165, 10)
plt.plot(x, cdf)
plt.xlabel("CDF")


plt.show()
```

In figure 1 and 2 are represented the Probability Density Function $f(x)$ and the Cumulative Distribution Function $F(x)$ respectively.
The CDF represent the same concept of the discrete example but in a continuous space.

Given a point on the CDF of coordinates $(x, y)$, $y$ is the probability of "getting" a woman height less than or equal to $x$.
Graphically, $y$ represent the area under the PDF curve delimited by a vertical line passing through $x$.

```python
plt.figure(figsize=(7,10))

# simulate a measurement of 100 women height, in a way that the mean is 165 and the standard deviation is 10
x = np.sort(np.random.normal(165, 10, 100))

plt.subplot(4,2,1)
height_dist = stats.norm.pdf(x, 165, 10)
plt.plot(x, height_dist)
plt.plot([160]*len(height_dist), height_dist)
plt.xlabel("PDF")

plt.subplot(4,2,2)
cdf = stats.norm.cdf(x, 165, 10)
plt.plot(x, cdf)
plt.scatter(160, stats.norm.cdf(160, 165, 10), color='g')
plt.xlabel("CDF")

plt.show()
```

Now that we have seen what the values of the CDF represent, we can focus on the values of the PDF. The PDF function values represent the gradient of the CDF. In other word the values represent the slope/inclination of the line with respect to the vertical axis and the horizontal axis.

Given two points $(x_1, y_1)$ and $(x_2, y_2)$, the gradient of the line, also called "Rise over the Run" is:
\begin{equation}
\frac{Rice}{Run} = \frac{(y_2-y_1)}{(x_2-x_1)}
\end{equation}

You can also calculate the gradient by differentiation:
\begin{equation}
f'(x) = \frac{dy}{dx}
\end{equation}

Moreover, the y-values of the PDF function represent the Likelihood

To sum up, to calculate the Probability Density Function from the Cumulative Distribution Function it is enough to calculate the derivative of the CDF. On the other hand, to calculate the CDF from the PDF you can compute the integral of the PDF.
\begin{equation}
CDF --> PDF = \frac{dF(x)}{dx} = f(x)
\end{equation}

\begin{equation}
PDF --> CDF = $\int_{-\infty}^x f(x)dx$ = F(x)
\end{equation}
